```python
# helper functions for displaying table data

import numpy as np
from IPython.display import display_html

# n is the number of columns to display data in
def display_side_by_side(series_obj, n):
    df = pd.DataFrame(series_obj)
    partition = int(round(len(df) / n))
    lower_bound = 0
    upper_bound = partition
    args = []
    for i in range(n):
        args.append(df[lower_bound:upper_bound])
        lower_bound += partition
        upper_bound += partition
    helper(args)

def helper(args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import csv

df = pd.read_csv('../data/OP_STIM_ML_JUN192020_imputed.csv')
df = df.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
df['SUDSy_0_cd'].value_counts()
```

```python
# group SUDSy_0_cd
df['SUDSy_0_cd'] += 10 # add 10 to prevent overlapping replacement

df['SUDSy_0_cd'] = np.where(df['SUDSy_0_cd'].between(10,11), 1, df['SUDSy_0_cd'])
df['SUDSy_0_cd'] = np.where(df['SUDSy_0_cd'].between(12,15), 2, df['SUDSy_0_cd'])
df['SUDSy_0_cd'] = np.where(df['SUDSy_0_cd'].between(16,22), 3, df['SUDSy_0_cd'])

df['SUDSy_0_cd'].value_counts()
```

```python
df = df[df.B2a_0 <= 29]
np.unique(df['B2a_0']).tolist() # confirm that this dataset contains adolescents AND young adults
```

```python
cols_to_use = ['ID','State','City','agyaddr','xobsyr_0','B2a_0','Illicit_Days','Illicit_Cens']

unimputed_cols_to_use = ['female','nonwhite','primsev','loc']

# keep imputed columns 
imputed_cols_to_use = ['unemplmt_cd','prsatx_cd','gvsg_cd','CWSg_0_cd',
                       'dssg_0_cd','epsg_0_cd','adhdg_0_cd','cdsg_0_cd','cjsig_0_cd','lrig_0_cd','srig_0_cd',
                       'SESg_0_cd','r4ag_0_cd','SUDSy_0_cd','homeless_0_cd','ncar_cd','TRIg_0_cd','PYS9Sxg_cd']

df = df[cols_to_use + unimputed_cols_to_use + imputed_cols_to_use]
df.head()
```

```python
df.shape
```

```python
from tqdm.notebook import tqdm

# ad hoc way of getting censoring to be correct
df['Illicit_Cens'] += 3 # add 3 to prevent overlapping replacement
# no relapse and censored data needs to be censored in the overall model
df.replace({'Illicit_Cens': {3: 1, 4: 0, 5:0}}, inplace=True) # 3 == relapsed, 4 == no relapse, 5 == censored

df['Illicit_Cens'].value_counts()
```

```python
df.replace(to_replace=['Yes', 'No'], value=[1, 0], inplace=True)
```

```python
cat_vars = ['primsev']

for var in cat_vars:
    # one-hot encode categorical variables
    one_hot = pd.get_dummies(df[var], prefix=var)
    df = df.drop(var, axis = 1)
    df = df.join(one_hot)

df.rename(columns={'primsev_Other drug':'primsev_other'}, inplace=True)
df.head()
```

```python
# create grouped age variable based on Mike's suggestions
# >15, 15-18, <18
df['B2a_0g'] = np.nan
for i,row in tqdm(df.iterrows(), total=df.shape[0]):
    age = row['B2a_0']
    if age < 15:
        df.set_value(i, 'B2a_0g', 0)
    elif age < 18:
        df.set_value(i, 'B2a_0g', 1)
    else:
        df.set_value(i, 'B2a_0g', 2)

df['B2a_0g'] = df['B2a_0g'].astype(int)
df.drop(columns=['B2a_0'], inplace=True)
df.head()
```

```python
# subset data based on whether the patient lived at the clinic or not
values = ['STR','MTR','LTR','Corrections']

df = df[~df['loc'].isin(values)]

df.drop(columns=['loc'], inplace=True)
df.shape
```

```python
# subtract 1 from imputed columns to keep levels same as rest of dataset
for col in df.columns:
    if '_cd' in col and col != 'SUDSy_0_cd':
        df[col] = df[col] - 1

df.head()
```
