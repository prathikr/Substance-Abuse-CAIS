```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import csv

df = pd.read_csv('../data/data_superset_temp.csv')
df.head()
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt

raw_census_vars = ['pop_deng', '%_dropoutg', '%_unemployedg', '%_public_assistanceg', '%_povertyg']

for i in raw_census_vars:
    plt.hist(df[i])
    plt.title(i + ' Pre-Filtering')
    plt.ylim(top=15000, bottom=0)
    plt.savefig('../graphs/'+i+'_pre_filtering.png')
    plt.show()
```
